## **4. Tách các cột có dữ liệu là chuỗi và tạo bảng dim cho các trường đó**

### **4.1. Tách cột belongs_to_collection trong bảng df_movies_metadata**

```python
# Parse cột belongs_to_collection thành dict
df_movies_metadata['belongs_to_collection'] = df_movies_metadata['belongs_to_collection'].apply(
    lambda x: ast.literal_eval(x) if pd.notnull(x) else {}
)

# Tạo collection_id trong bảng fact (dùng cho fact_movies_metadata nếu cần)
df_movies_metadata['collection_id'] = df_movies_metadata['belongs_to_collection'].apply(
    lambda x: x.get('id') if isinstance(x, dict) else None
)

# Tạo bảng dim_collections (lấy id, name, poster_path, backdrop_path)
dim_collections = df_movies_metadata['belongs_to_collection'].apply(
    lambda x: {
        'id': x.get('id'),
        'name': x.get('name'),
        'poster_path': x.get('poster_path'),
        'backdrop_path': x.get('backdrop_path')
    } if isinstance(x, dict) and 'id' in x else None
)

# Loại bỏ None và trùng lặp
dim_collections = pd.DataFrame([d for d in dim_collections if d is not None]).drop_duplicates().reset_index(drop=True)

# Đổi tên cột cho rõ nghĩa
dim_collections = dim_collections.rename(columns={
    'id': 'collection_id',
    'name': 'collection_name'
})

# Xóa cột gốc
df_movies_metadata = df_movies_metadata.drop(columns=['belongs_to_collection'])

# Lưu ra file
dim_collections.to_csv(r'C:\FPT Polytechnic\Project Tự Làm\TMDb\Data\Processed\dim_collections.csv', index=False, encoding='utf-8-sig')
```

```python
df_movies_metadata.columns
```

### **4.2. Tách cột Genres trong bảng df_movies_metadata**

```python
df_movies_metadata['genres'] = df_movies_metadata['genres'].apply(
    lambda x: ast.literal_eval(x) if pd.notnull(x) else []
)

df_movies_metadata['genre_id'] = df_movies_metadata['genres'].apply(
    lambda genre_list: ', '.join([str(genre['id']) for genre in genre_list]) if isinstance(genre_list, list) else None
)

all_genres = df_movies_metadata['genres'].explode()
dim_genres = pd.DataFrame([g for g in all_genres if isinstance(g, dict)])
dim_genres = dim_genres[['id', 'name']].drop_duplicates().reset_index(drop=True)

df_movies_metadata = df_movies_metadata.drop(columns=['genres'])

dim_genres.to_csv(r'C:\FPT Polytechnic\Project Tự Làm\TMDb\Data\Processed\dim_genres.csv', index=False, encoding='utf-8-sig')
```

```python
df_movies_metadata.columns
```

### **4.3. Tách cột production_companies trong bảng df_movies_metadata**

```python
df_movies_metadata['production_companies'] = df_movies_metadata['production_companies'].apply(
    lambda x: ast.literal_eval(x) if pd.notnull(x) else []
)

df_movies_metadata['company_id'] = df_movies_metadata['production_companies'].apply(
    lambda comps: ', '.join([str(comp['id']) for comp in comps]) if isinstance(comps, list) else None
)

all_companies = df_movies_metadata['production_companies'].explode()
dim_production_companies = pd.DataFrame([c for c in all_companies if isinstance(c, dict)])
dim_production_companies = dim_production_companies[['id', 'name']].drop_duplicates().reset_index(drop=True)

df_movies_metadata = df_movies_metadata.drop(columns=['production_companies'])

dim_production_companies.to_csv(r'C:\FPT Polytechnic\Project Tự Làm\TMDb\Data\Processed\dim_production_companies.csv', index=False, encoding='utf-8-sig')
```

### **4.4. Tách cột production_countries trong bảng df_movies_metadata**

```python
df_movies_metadata['production_countries'] = df_movies_metadata['production_countries'].apply(
    lambda x: ast.literal_eval(x) if pd.notnull(x) else []
)

df_movies_metadata['country_code'] = df_movies_metadata['production_countries'].apply(
    lambda countries: ', '.join([c['iso_3166_1'] for c in countries]) if isinstance(countries, list) else None
)

all_countries = df_movies_metadata['production_countries'].explode()
dim_production_countries = pd.DataFrame([c for c in all_countries if isinstance(c, dict)])
dim_production_countries = dim_production_countries[['iso_3166_1', 'name']].drop_duplicates().reset_index(drop=True)
dim_production_countries.columns = ['country_code', 'country_name']

df_movies_metadata = df_movies_metadata.drop(columns=['production_countries'])

dim_production_countries.to_csv(r'C:\FPT Polytechnic\Project Tự Làm\TMDb\Data\Processed\dim_production_countries.csv', index=False, encoding='utf-8-sig')
```

### **4.5. Tách cột spoken_language trong bảng df_movies_metadata**

```python
df_movies_metadata['spoken_languages'] = df_movies_metadata['spoken_languages'].apply(
    lambda x: ast.literal_eval(x) if pd.notnull(x) else []
)

df_movies_metadata['language_code'] = df_movies_metadata['spoken_languages'].apply(
    lambda langs: ', '.join([lang['iso_639_1'] for lang in langs]) if isinstance(langs, list) else None
)

all_langs = df_movies_metadata['spoken_languages'].explode()
dim_spoken_languages = pd.DataFrame([l for l in all_langs if isinstance(l, dict)])
dim_spoken_languages = dim_spoken_languages[['iso_639_1', 'name']].drop_duplicates().reset_index(drop=True)
dim_spoken_languages.columns = ['language_code', 'language_name']

df_movies_metadata = df_movies_metadata.drop(columns=['spoken_languages'])

dim_spoken_languages.to_csv(r'C:\FPT Polytechnic\Project Tự Làm\TMDb\Data\Processed\dim_spoken_languages.csv', index=False, encoding='utf-8-sig')
```

### **4.6. Chuẩn hóa bảng credits**

```python
def parse_json_column(col):
    return col.apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])

# Parse cột cast và crew
df_credits['cast'] = parse_json_column(df_credits['cast'])
df_credits['crew'] = parse_json_column(df_credits['crew'])

# Chuẩn hóa dữ liệu cast
cast_records = []
for _, row in df_credits.iterrows():
    for person in row['cast']:
        cast_records.append({
            'movie_id': row['id'],
            'person_id': person.get('id'),
            'character': person.get('character'),
            'role': 'cast'
        })

# Chuẩn hóa dữ liệu crew
crew_records = []
for _, row in df_credits.iterrows():
    for person in row['crew']:
        crew_records.append({
            'movie_id': row['id'],
            'person_id': person.get('id'),
            'character': None,
            'role': person.get('job')  # Ví dụ: 'Director', 'Producer', ...
        })

# Gộp lại thành bảng fact
fact_movies_credits = pd.DataFrame(cast_records + crew_records)

# Tạo dim_person (loại trùng lặp)
persons = set()
for lst in df_credits['cast']:
    persons.update([(p['id'], p['name']) for p in lst])
for lst in df_credits['crew']:
    persons.update([(p['id'], p['name']) for p in lst])

dim_person = pd.DataFrame(list(persons), columns=['person_id', 'person_name'])

# Tạo dim_role từ cột role duy nhất
dim_role = pd.DataFrame(fact_movies_credits['role'].dropna().unique(), columns=['role_name'])
dim_role['role_id'] = dim_role.index + 1

# Map role_id vào fact table
fact_movies_credits = fact_movies_credits.merge(dim_role, left_on='role', right_on='role_name', how='left')
fact_movies_credits = fact_movies_credits[['movie_id', 'person_id', 'role_id', 'character']]

# Lưu ra file CSV
fact_movies_credits.to_csv(r"C:\FPT Polytechnic\Project Tự Làm\TMDb\Data\Processed\fact_movie_credits.csv", index=False)
dim_person.to_csv(r"C:\FPT Polytechnic\Project Tự Làm\TMDb\Data\Processed\dim_person.csv", index=False)
dim_role.to_csv(r"C:\FPT Polytechnic\Project Tự Làm\TMDb\Data\Processed\dim_role.csv", index=False)
```

### **4.8. Chuẩn hóa bảng keywords**

```python
# Chuyển cột keywords từ string sang list
df_keywords['keywords'] = df_keywords['keywords'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])

# Tạo bảng dim_keyword
dim_keyword_set = set()
for keyword_list in df_keywords['keywords']:
    for kw in keyword_list:
        dim_keyword_set.add((kw['id'], kw['name']))

df_dim_keyword = pd.DataFrame(list(dim_keyword_set), columns=['keyword_id', 'keyword_name'])

# Tạo bảng fact_movie_keywords
fact_movie_keyword_links = []
for _, row in df_keywords.iterrows():
    movie_id = row['id']
    for kw in row['keywords']:
        fact_movie_keyword_links.append((movie_id, kw['id']))

df_movie_keywords = pd.DataFrame(fact_movie_keyword_links, columns=['movie_id', 'keyword_id'])

# Xuất file
df_dim_keyword.to_csv(r"C:\FPT Polytechnic\Project Tự Làm\TMDb\Data\Processed\dim_keyword.csv", index=False)
df_movie_keywords.to_csv(r"C:\FPT Polytechnic\Project Tự Làm\TMDb\Data\Processed\fact_movie_keywords.csv", index=False)
```
